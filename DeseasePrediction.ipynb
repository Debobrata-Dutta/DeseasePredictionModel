```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Step 1: Create the corrected dataset
data = {
    'fever': ['yes', 'no', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'no'],
    'cough': ['yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'no', 'yes', 'yes'],
    'fatigue': ['no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes'],
    'age': ['young', 'adult', 'adult', 'senior', 'senior', 'adult', 'young', 'adult', 'senior', 'adult'],
    'disease': ['yes', 'no', 'yes', 'yes', 'no', 'yes', 'no', 'no', 'yes', 'no']
}

df = pd.DataFrame(data)

# Step 2: Encode categorical data
le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])

# Step 3: Split data into features (X) and target (y)
X = df.drop('disease', axis=1)
y = df['disease']

# Step 4: Build the Decision Tree using Gini index
clf = DecisionTreeClassifier(criterion='gini')
clf = clf.fit(X, y)
accuracy = clf.score(X, y)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Step 5: Visualize the Decision Tree
plt.figure(figsize=(18,8))
plot_tree(clf, filled=True, feature_names=['fever', 'cough', 'fatigue', 'age'], class_names=['no', 'yes'])
plt.show()
```
